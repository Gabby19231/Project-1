```python
max_votes = states_DR.groupby(['year','state'])['candidatevotes'].max()
maxDF = pd.DataFrame(max_votes)
maxDF
#Dataframe with max votes by year by state
```

```python
winner = maxDF.merge(states_DR, how='inner', left_on=["candidatevotes"], right_on=["candidatevotes"])
winner
#merging to have a dataframe with the candidate with majority of the votes of each election by state
```

```python
winner2 = winner.groupby(['year','state','candidate'])['candidatevotes'].sum()
winner2
```

```python
#winner = ['Carter, Jimmy', 'Reagan, Ronald', 'Reagan, Ronald', 'Bush, George H.W.', 'Clinton, Bill', 'Clinton, Bill', 'Bush, George W.', 'Bush, George W.', 'Obama, Barack H.', 'Obama, Barack H.','Trump, Donald J.']
JC = winner.loc[(winner['candidate'] == 'Carter, Jimmy') & (winner['year'] == 1976)]
RR = winner.loc[(winner['candidate'] == 'Reagan, Ronald') & (winner['year'] == 1980) | (winner['candidate'] == 'Reagan, Ronald') & (winner['year'] == 1984)]
GB1 = winner.loc[(winner['candidate'] == 'Bush, George H.W.') & (winner['year'] == 1988)]
BC = winner.loc[(winner['candidate'] == 'Clinton, Bill') & (winner['year'] == 1992) | (winner['candidate'] == 'Clinton, Bill') & (winner['year'] == 1996)]
GB2 = winner.loc[(winner['candidate'] == 'Bush, George W.') & (winner['year'] == 2000) | (winner['candidate'] == 'Bush, George W.') & (winner['year'] == 2004)]
BO = winner.loc[(winner['candidate'] == 'Obama, Barack H.') & (winner['year'] == 2008) | (winner['candidate'] == 'Obama, Barack H.') & (winner['year'] == 2012)]
DT = winner.loc[(winner['candidate'] == 'Trump, Donald J.') & (winner['year'] == 2016)]

test = JC.append(RR)
test2 = test.append(GB1)
test3 = test2.append(BC)
test4 = test3.append(GB2)
test5 = test4.append(BO)
test6 = test5.append(DT)

test6
#DataFrame showing in wich states the elected presidents had the majority of the votes
```

```python
plt.figure(figsize=(10,8))
states_count = test6['state'].value_counts()
plt.bar(states_count.index.values, states_count.values)
plt.xticks(rotation=90)
plt.title('Number of time the elected candidate had the majority of the votes by state')
plt.show()
#Number of times the popular votes in the state predicted the presidencial election winner
#Ohio, FLorida, Pennsylvania
```
