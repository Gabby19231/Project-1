```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
path = "projections_2020_asrh.xlsx - Proj2019-2045.csv"

# Read in Florida census data
florida = pd.read_csv(path)

# Clean and reformat Florida data
florida = florida.transpose()
florida = florida.rename(columns = {0: "all_total", 
                                1: "all_female", 
                                2: "all_male", 
                                3: "nh_white_total", 
                                4: "nh_white_female", 
                                5: "nh_white_male",
                                6: "nh_black_total", 
                                7: "nh_black_female", 
                                8: "nh_black_male",
                                9: "hispanic_total", 
                                10: "hispanic_female", 
                                11: "hispanic_male",
                                12: "other_total", 
                                13: "other_female", 
                                14: "other_male"})

# Display the data table for preview
florida.head()
```

```python
# Read in voter data and display
# 2016 voter breakdown by political party. Based on CNN exit polls

voter_path = "voter_breakdown_2016 - florida.csv"

race_voter = pd.read_csv(voter_path)
race_voter = race_voter.groupby(['race','gender']).sum()
race_voter
```

```python
from turnout import turnout

turnout
```

```python
# Clean data for plotting racial data over time
# No breakdown by age or gender

demographic_data = florida.drop(index = ['Race', 'Sex'])
years = [2010, 2019, 2020, 2025, 2030, 2035, 2040, 2045]

# demographic_data = demographic_data.set_index('year')
total_pop = demographic_data['all_total']

total_df = demographic_data[['nh_white_total','nh_black_total','hispanic_total','other_total']]


# Plot race over time
total_df.plot(title = 'Florida Racial Demographic Changes over Time')
plt.ylabel('Population')
plt.xlabel('Year')
plt.show()

total_df.head()
```

```python
demographic_data.head()
```

```python
rep_breakdown = race_voter['trump'] / 100
rate = turnout['voting_rate'] / 100

rep_white_female = (demographic_data['nh_white_female'] * rate[6] * rep_breakdown[7]).tolist()
rep_white_male = (demographic_data['nh_white_male'] * rate[7] * rep_breakdown[6]).tolist()
rep_black_female = (demographic_data['nh_black_female'] * rate[2] * rep_breakdown[1]).tolist()
rep_black_male = (demographic_data['nh_black_male'] * rate[3] * rep_breakdown[0]).tolist()
rep_hispanic_female = (demographic_data['hispanic_female'] * rate[4] * rep_breakdown[3]).tolist()
rep_hispanic_male = (demographic_data['hispanic_male'] * rate[5] * rep_breakdown[2]).tolist()
rep_asian_female = (demographic_data['other_female'] * rate[0] * rep_breakdown[5]).tolist()
rep_asian_male = (demographic_data['other_male'] * rate[1] * rep_breakdown[4]).tolist()

rep_total = []
for i in range(len(years)):
    rep_total.append(rep_white_female[i] + 
                     rep_black_female[i] + 
                     rep_hispanic_female[i] + 
                     rep_asian_female[i] + 
                     rep_white_male[i] + 
                     rep_black_male[i] + 
                     rep_hispanic_male[i] + 
                     rep_asian_male[i])

rep_forecast_df = pd.DataFrame({
    'White Women': rep_white_female,
    'White Men': rep_white_male,
    'Black Women': rep_black_female,
    'Black Men': rep_black_male,
    'Hispanic Women': rep_hispanic_female,
    'Hispanic Men': rep_hispanic_male,
    'Asian Women': rep_asian_female,
    'Asian Men': rep_asian_male
   }, index=years)

rep_forecast_df.head()
```

```python
dem_breakdown = race_voter['clinton'] / 100

dem_white_female = (demographic_data['nh_white_female'] * rate[6] * dem_breakdown[7]).tolist()
dem_white_male = (demographic_data['nh_white_male'] * rate[7] * dem_breakdown[6]).tolist()
dem_black_female = (demographic_data['nh_black_female'] * rate[2] * dem_breakdown[1]).tolist()
dem_black_male = (demographic_data['nh_black_male'] * rate[3] * dem_breakdown[0]).tolist()
dem_hispanic_female = (demographic_data['hispanic_female'] * rate[4] * dem_breakdown[3]).tolist()
dem_hispanic_male = (demographic_data['hispanic_male'] * rate[5] * dem_breakdown[2]).tolist()
dem_asian_female = (demographic_data['other_female'] * rate[0] * dem_breakdown[5]).tolist()
dem_asian_male = (demographic_data['other_male'] * rate[1] * dem_breakdown[4]).tolist()

dem_total = []
for i in range(len(years)):
    dem_total.append(dem_white_female[i] + 
                     dem_black_female[i] + 
                     dem_hispanic_female[i] + 
                     dem_asian_female[i] + 
                     dem_white_male[i] + 
                     dem_black_male[i] + 
                     dem_hispanic_male[i] + 
                     dem_asian_male[i])

dem_forecast_df = pd.DataFrame({
    'White Women': dem_white_female,
    'White Men': dem_white_male,
    'Black Women': dem_black_female,
    'Black Men': dem_black_male,
    'Hispanic Women': dem_hispanic_female,
    'Hispanic Men': dem_hispanic_male,
    'Asian Women': dem_asian_female,
    'Asian Men': dem_asian_male
   }, index=years)

dem_forecast_df.head()
```

```python
forecast_df = pd.DataFrame({
   'Dem': dem_total,
   'Rep': rep_total
   }, index=years)

forecast_df.head()

plt.plot(years,dem_total, label = "Dem", color = 'b')
plt.plot(years,rep_total, label = "Rep", color = 'r')
plt.ylabel('# of Votes')
plt.xlabel('Year')
plt.title('Florida Political Projection')
plt.legend(loc='best')
plt.show()
```

```python
# Same racial demographics as above, but as % of total population, instead of population totals. 

# demographic_data_percent = total_df

white_percent = (total_df['nh_white_total']/total_pop).tolist()
black_percent = (total_df['nh_black_total']/total_pop).tolist()
hispanic_percent = (total_df['hispanic_total']/total_pop).tolist()
asian_percent = (total_df['other_total']/total_pop).tolist()


demographic_data_percent = pd.DataFrame({
    'White': white_percent,
    'Black': black_percent,
    'Hispanic': hispanic_percent,
    'Asian': asian_percent
   }, index=years)


demographic_data_percent.plot(title = 'Florida Racial Group as Proportion of Total Population over Time')
plt.ylabel('% of Total Population')
plt.xlabel('Year')
plt.show()

demographic_data_percent.head()
```
