```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
path = "State_of_Texas.csv"

# Read in Texas census data
texas = pd.read_csv(path)
texas = texas.drop(columns=['FIPS', 'area_name'])

# Display the data table for preview
texas.head()
```

```python
# Read in voter data and display
# 2016 voter breakdown by political party. Based on CNN exit polls

voter_path = "voter_breakdown_2016 - texas.csv"

texas_race_voter = pd.read_csv(voter_path)
texas_race_voter = texas_race_voter.groupby(['race','gender']).sum()
texas_race_voter
```

```python
from turnout import turnout

turnout
```

```python
# Clean data for plotting racial data over time
# No breakdown by age or gender

demographic_data = texas[texas['age_group'] == 'ALL']
years = demographic_data['year']

demographic_data = demographic_data.set_index('year')
total_pop = demographic_data['total']

total_df = demographic_data[['nh_white_total','nh_black_total','hispanic_total','nh_asian_total']]


# Plot race over time
total_df.plot(title = 'Texas Racial Demographic Changes over Time')
plt.ylabel('Population')
plt.show()

total_df.head()
```

```python
demographic_data.head()
```

```python
rep_breakdown = texas_race_voter['trump'] / 100
rate = turnout['voting_rate'] / 100

rep_white_female = demographic_data['nh_white_female']  * rep_breakdown[7] * rate[6]
rep_white_male = demographic_data['nh_white_male']  * rep_breakdown[6] * rate[7]
rep_black_female = demographic_data['nh_black_female'] * rep_breakdown[1] * rate[2] 
rep_black_male = demographic_data['nh_black_male'] * rep_breakdown[0] * rate[3] 
rep_hispanic_female = demographic_data['hispanic_female']  * rep_breakdown[3] * rate[4]
rep_hispanic_male = demographic_data['hispanic_male']  * rep_breakdown[2] * rate[5]
rep_asian_female = demographic_data['nh_asian_female']  * rep_breakdown[5] * rate[0]
rep_asian_male = demographic_data['nh_asian_male']  * rep_breakdown[4] * rate[1]

rep_total = rep_white_female + rep_black_female + rep_hispanic_female + rep_asian_female + rep_white_male + rep_black_male + rep_hispanic_male + rep_asian_male

rep_forecast_df = pd.DataFrame({
    'White Women': rep_white_female,
    'White Men': rep_white_male,
    'Black Women': rep_black_female,
    'Black Men': rep_black_male,
    'Hispanic Women': rep_hispanic_female,
    'Hispanic Men': rep_hispanic_male,
    'Asian Women': rep_asian_female,
    'Asian Men': rep_asian_male
   }, index=years)

rep_forecast_df.head()
```

```python
dem_breakdown = texas_race_voter['clinton'] / 100

dem_white_female = demographic_data['nh_white_female']  * dem_breakdown[7] * rate[6]
dem_white_male = demographic_data['nh_white_male']  * dem_breakdown[6] * rate[7]
dem_black_female = demographic_data['nh_black_female']  * dem_breakdown[1] * rate[2]
dem_black_male = demographic_data['nh_black_male']  * dem_breakdown[0] * rate[3]
dem_hispanic_female = demographic_data['hispanic_female']  * dem_breakdown[3] * rate[4]
dem_hispanic_male = demographic_data['hispanic_male']  * dem_breakdown[2] * rate[5]
dem_asian_female = demographic_data['nh_asian_female']  * dem_breakdown[5] * rate[0]
dem_asian_male = demographic_data['nh_asian_male'] * dem_breakdown[4] * rate[1] 

dem_total = dem_white_female + dem_black_female + dem_hispanic_female + dem_asian_female + dem_white_male + dem_black_male + dem_hispanic_male + dem_asian_male

dem_forecast_df = pd.DataFrame({
    'White Women': dem_white_female,
    'White Men': dem_white_male,
    'Black Women': dem_black_female,
    'Black Men': dem_black_male,
    'Hispanic Women': dem_hispanic_female,
    'Hispanic Men': dem_hispanic_male,
    'Asian Women': dem_asian_female,
    'Asian Men': dem_asian_male
   }, index=years)

dem_forecast_df.head()
```

```python
forecast_df = pd.DataFrame({
   'Dem': dem_total,
   'Rep': rep_total
   }, index=years)

forecast_df.head()

plt.plot(years,dem_total, label = "Dem", color = 'b')
plt.plot(years,rep_total, label = "Rep", color = 'r')
plt.ylabel('# of Votes')
plt.xlabel('Year')
plt.title('Texas Political Projection')
plt.legend(loc='best')
plt.show()
```

```python
# Same racial demographics as above, but as % of total population, instead of population totals. 

# demographic_data_percent = total_df

white_percent = total_df['nh_white_total']/total_pop
black_percent = total_df['nh_black_total']/total_pop
hispanic_percent = total_df['hispanic_total']/total_pop
asian_percent = total_df['nh_asian_total']/total_pop


demographic_data_percent = pd.DataFrame({
    'White': white_percent,
    'Black': black_percent,
    'Hispanic': hispanic_percent,
    'Asian': asian_percent
   }, index=years)


demographic_data_percent.plot(title = 'Texas Racial Group as Proportion of Total Population over Time')
plt.ylabel('% of Total Population')
plt.show()

demographic_data_percent.head()
```

```python
# Clean data for plotting age data over time
# No breakdown by race or gender

age_data = texas[['year','age_group','total']]
age_data = pd.pivot_table(age_data, values = 'total', index = 'year', columns = 'age_group')
total_pop_age = age_data['ALL']
age_data = age_data.drop(columns=['ALL'])


# Plot age over time
age_data.plot(title = 'Texas Age Group Projection over Time')
plt.ylabel('Population')
plt.show()
```

```python
# Same age demographics as above, but as % of total population, instead of population totals. 

age_data_percent = age_data

age_data_percent['18-24'] = age_data_percent['18-24']/total_pop_age
age_data_percent['25-44'] = age_data_percent['25-44']/total_pop_age
age_data_percent['45-64'] = age_data_percent['45-64']/total_pop_age
age_data_percent['65+'] = age_data_percent['65+']/total_pop_age
age_data_percent['<18'] = age_data_percent['<18']/total_pop_age


age_data_percent.plot(title = 'Texas Age Group as Proportion of Total Population over Time')
plt.ylabel('% of Total Population')
plt.show()
```
